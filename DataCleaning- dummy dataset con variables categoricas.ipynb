```python
import pandas as pd
import numpy as np
```

```python
gender = ["Male" ,"Female"] #genero
income = ["Poor", "Middle Class", "Rich"] #ingresos
```

```python
n=500
gender_data = [] #arreglo de genero vacio
income_data = [] #arreglo de ingresos vacio
i=0
for i in range(n):
    gender_data.append(np.random.choice(gender)) ##el arreglo "gender_data" se rellena aleatoriamente con los datos del arreglo "gender" 
    income_data.append(np.random.choice(income)) ##el arreglo "income_data" se rellena aleatoriamente con los datos del arreglo "income" 
```

```python
height = 160 + 30 + np.random.randn(n) ##se generan aleatoriamente n valores con una distribución normal con media=160 y desvición=30
weight = 65 + 25 + np.random.randn(n)  ##se generan aleatoriamente n valores con una distribución normal con media=65 y desvición=25
age = 30 + 12 * np.random.randn(n)     ##se generan aleatoriamente n valores con una distribución normal con media=30 y desvición=12
income = 18000 + 3500 * np.random.randn(n) ##se generan aleatoriamente n valores con una distribución normal con media=18000 y desvición=3500
```

```python
data = pd.DataFrame(        ## se genera el dummy dataframe con los valores generados anteriormente
    {
        "Gender": gender_data,
        "Economic Status": income_data,
        "height" : height,
        "weight": weight,
        "age": age,
        "income": income
    }
)
```

```python
data.head()
```

### agrupación de datos

```python
##agrupación por ua sola categoría. En este caso por la categoria "Gender"
grupo_gender = data.groupby("Gender")
```

```python
grupo_gender.groups
```

```python
for names, groups in grupo_gender:
    print(names)
    print(groups)
```

```python
## para obterner un solo grupo la sentencia es la sig:
grupo_gender.get_group("Female")
```

```python
## agrupación por mas de una categoría. En este caso por la categoria "Gender" y "Economi Status"
grupo_doble = data.groupby(["Gender", "Economic Status"])
```

```python
for names, groups in grupo_doble:
    print(names)
    print(groups)
```

# operaciones sobre los datos agrupados

```python
grupo_doble.sum() ## realiza una suma de las alturas,los pesos ,las edades y el salario de cada grupo
```

```python
grupo_doble.mean() ##realiza el promedio de las alturas,los pesos, las edades y el salario de cada grupo
```

```python
grupo_doble.size() ##devuelve la cantidad de elementos que hay en cada una de las categorias
```

```python
grupo_doble.describe() ## describe el valor mínimo, el valor máximo, el promedio, la desviación de cada una de las variables.
```

### para que sea más legible, se puede obtener sólo una de las columnas y sobre ella hacer las operaciones:

```python
grupo_doble["income"].describe() ## en este caso se obtuvo solo la columna ingresos
```

### las funciones de agregado también se las puede hacer de una manera más personalizada

```python
grupo_doble.aggregate(
    {
        "income":np.sum, ##suma los ingresos
        "age": np.mean, ##promedio de la edades
        "height":np.std ##desviación de las alturas
    }
    
)
```

### para hacer las funciones de agregado aún más personalizadas se puede aplicar una función lambda que me permite hacer la operación que quiera. por ejemplo:

```python
grupo_doble.aggregate(
    {
        "income": lambda ingresos:(np.sum(ingresos) / 12), ##en este caso saco el promedio de los ingresos por mes al hacer la divicion entre 12
        "age": np.mean ##obtiene los promedios de las edades
    }
)
```
