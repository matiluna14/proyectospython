```python
import tensorflow as tf
import matplotlib.pyplot as plt
session = tf.Session()
```

# TIPOS DE FUNCIONES DE PERDIDA EN PROBLEMAS DE REGRESION (problemas que comprenden valores continuos)

```python
x_vals = tf.linspace(-1.,1.,500) ##genera un conjunto de 500 valores continuos entre (-1 y 1) que actuaran como valores predecidos
target = tf.constant(0.) ## actua como los valores reales
```

```python
print(session.run(x_vals))
```

```python
print(session.run(target))
```

# norma en L2 (distancia Euclidea)

### L2 =  (yr - yp)^2

```python
l2 = tf.square(target - x_vals)
```

```python
print(session.run(l2))
```

# norma en L1 (valor absoluto)

### L1 = |yr - yp|

```python
L1 = tf.abs(target-x_vals)
```

```python
print(session.run(L1))
```

# Pseudo-Huber (combina lo mejor de la norma L2 y la norma L1)

### PH = ð^2(√1+(yr -yp) / ð)^2 - 1

```python
delta = tf.constant(0.25)
ph = tf.multiply(tf.square(delta),tf.sqrt(1.+tf.square((target-x_vals)/delta))-1.)
```

```python
print(session.run(ph))
```

# TIPOS DE FUNCIONES DE PERDIDA EN PROBLEMAS DE CLASIFICACION
    -cuanto mas cercano a 0 se encuentre el valor del error, mejor es la prediccion; y cuanto mas cercano a 1 peor es la prediccion

```python
x_vals = tf.linspace(-3.,5.,500) ##genera un conjunto de 500 valores continuos entre (-3 y 5) que actuaran como valores predecidos
target = tf.constant(1.) 
targets =tf.fill([500,],1.) ## dimension de 500 valores 1. actua como los valores reales.
```

```python
print(session.run(x_vals))
```

```python
print(session.run(target))
```

# Hinge (funcion bisagra)

### Hinge = max(0,1-(yr*yp))

```python
hinge = tf.maximum(0., 1.-tf.multiply(target,x_vals))
```

```python
print(session.run(hinge))
```

# Entropia cruzada (funcion logistica)
    -funciona muy bien cuando se trata de predecir 2 clases binarias (osea 0 o 1)

### H = -yr * log(yp) - (1-yr) * log(1-yp)

```python
cross_entropy = -tf.multiply(target, tf.log(x_vals)) - tf.multiply(1.-target, tf.log(1.-x_vals))
```

```python
print(session.run(cross_entropy))
```

# Entropia cruzada de sigmoide

### H = -yr * log( (1/1+e^-yp) ) - (1-yr) * log( 1- (1/1+e^-yp) )

```python
cross_entropy = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_vals,labels=targets)
```

```python
print(session.run(cross_entropy))
```

# Softmax Cross Entropy 
    -funciona muy bien cuando se trata de predecir una unica clase.

```python
logitsposibles = tf.constant([[1.,-3.,10.]]) ##3 posibles categorias
target_dist = tf.constant([[0.1,0.02,0.88]]) ## variable a predecir, son probabilidades del propio algoritmo de pertenecer a una u otra categoria
softmax_cross_entropy = tf.nn.softmax_cross_entropy_with_logits_v2(logits=logitsposibles, labels=target_dist)
```

```python
print(session.run(softmax_cross_entropy))
```

# Sparse softmax cross entropy
    --funciona muy bien cuando se trata de predecir una unica clase.

```python
logitsposibles = tf.constant([[1.,-3.,10.]]) ##3 posibles categorias
sparsed_target_dist = tf.constant([2]) ## variable a predecir,es un vector que contiene la posicion de la categoria correcta
sparced_softmax_cross_entropy = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logitsposibles, labels=sparsed_target_dist)
```

```python
print(session.run(sparced_softmax_cross_entropy))
```
