```python
import tensorflow as tf
session = tf.Session()
```

```python
import numpy as np
```

```python
my_array = np.array([[1.,3.,5.,7.],
                   [-2.,0.,7.,6.],
                   [-6.,-1.,0.,3.]])
my_array
```

```python
x_vals = np.array([my_array])
```

## se definen los placeholders y las constantes

```python
x_data = tf.placeholder(tf.float32, shape=(3,4)) ##placeholder de 3 filas y 4 columnas
```

```python
m1 = tf.constant([[1.],[0.],[-2.],[5.]]) ##vector de 4 filas y 1 columna
m2 = tf.constant([[7.]]) ##vector de 1 fila y 1 columna
a1 = tf.constant([[15.]]) #vector de 1 fila y 1 columna
```

## la operacion a llevar a cabo va a ser:
    - [ (X_data * m1) m2] + a1

```python
## se define la operacion 
prod1 = tf.matmul(x_data, m1)
prod2 = tf.matmul(prod1, m2)
add1 = tf.add(prod2, a1)
```

```python
## se lleva a cabo la ejecucion de la operacion
for x_val in x_vals:
    print(session.run(add1, feed_dict={x_data: x_val}))
```
