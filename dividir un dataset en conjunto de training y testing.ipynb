```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Customer Churn Model.txt")
```

```python
cantFilasDataset=len(data) ##cantidad de filas del dataset
```

### dividir el dataset utilizando la distribución normal con la regla del 80-20 (el 80%de los datos para training y el 20% de los datos para testing)

```python
## se crea un vector con la misma longitud de nuestro dataset y se rellena con valores aleatorios con una distribución normal
vector = np.random.randn(cantFilasDataset) 
```

```python
plt.hist(vector) ## el vector creado con valores aleatorios sigue una distribución normal
```

```python
check = (vector<0.8) ## se crea un vector "check" con valores true y false, según se cumpla la condición. los valores true van a ser el 80% y los valores false van a ser el 20%
```

```python
check
```

```python
training = data[check] ##el 80% del dataset
training
```

```python
testing = data[~check] ##el 20% del dataset
testing
```

### dividir el dataset para training y testing con la librería sklearn y el método shuffle 

```python
import sklearn
```

```python
data = sklearn.utils.shuffle(data) ##mezcla el dataset aleatoriamente
```

```python
cut_id = int(0.8 * len(data)) ##se genera el "id de corte". 
train_data = data[:cut_id]    ## se corta el dataset desde la primera fila hasta el 80%
test_data = data[cut_id:]     ## se corta el dataset desde la el 80% hasta la última fila
```

```python
len(train_data) ##el 80% del dataset
```

```python
len(test_data) ##el 20% del dataset
```
