# creación de nuevas columnas para rellenar con los datos que queramos

# generación de números aleatorios

```python
import numpy as np
```

```python
## generar un número aleatorio entre 1 y 100
np.random.randint(1,100)
```

```python
## la forma más clásica de generar un número aleatorio es entre 0 y 1 (con decimales)
np.random.random()
```

```python
## función que genera un lista de n números aleatorios enteros en el intervalo [a,b]
def generadorNumerosAleatorios (n,a,b):
    listaAleatorios = []
    for i in range(n):
        listaAleatorios.append(np.random.randint(a,b))
    return listaAleatorios
```

```python
generadorNumerosAleatorios(12,0,1000)
```

```python
## generación de números aleatorios del 0 al 100 múltiplos de 7
import random 
random.randrange(0,100,7)
```

```python
## función para generar un lista de n números aleatorios enteros múltiplos de un valor en el intervalo [a,b]
def generadorNumerosAleatoriosMultiplos(n,a,b,m):
    listaAleatorios = []
    for i in range(n):
        listaAleatorios.append(random.randrange(a,b,m))
    return listaAleatorios

        
```

```python
generadorNumerosAleatoriosMultiplos(14,0,100,7)
```
