```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
import numpy as np
```

```python
data = pd.read_csv("Admission_Predict_Ver1.1.csv")
```

```python
data
```

```python
columns_names = data.columns.values.tolist()
columns_names
```

```python
descartar = ["Chance of Admit ","Serial No."]
columns_names = [v for v in data if v not in descartar]
columns_names
```

```python
X = data[columns_names]
Y = data["Chance of Admit "]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X,Y)
```

```python
lm.score(X,Y)
```

```python
lm.intercept_
```

```python
list(zip(columns_names,lm.coef_))
```

```python
data
```

```python
## se aÃ±ade al dataset la columna de "RSE" error estÃ¡ndar residual
data["RSE"] = (data["Chance of Admit "] - data["chance admit prediccion"])**2

##suma de los cuadrados de los errores
SSD = sum(data["RSE"])
print("la suma de los cuadrados de la diferencia es ",SSD)

## error estÃ¡ndar residual
RSE = np.sqrt(SSD/(len(data)-7-1))
print("error estÃ¡ndar residual es: ",RSE)

## promedio de ventas del dataset original
promedio_chanceAdmit = np.mean(data["Chance of Admit "])
print("el promedio de chanceAdmit del dataset original es: ",promedio_chanceAdmit)

##error del modelo
error = RSE/promedio_chanceAdmit
print("el error del modelo es: ",error)
```
