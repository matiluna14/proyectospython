### conversiÃ³n de las variables categÃ³ricas a dummies

```python
categories = ["job","marital","education","housing","loan","month","day_of_week","poutcome","default","contact"]

for category in categories: ##se convierten en dummies todas las columnas de "categories"
    cat_list = "cat"+"_"+category
    cat_dummies = pd.get_dummies(data[category], prefix=cat_list)
    data = data.join(cat_dummies)
```

### Curva ROC hecha a mano

```python
threshold = [0.04, 0.05 ,0.07, 0.10, 0.12, 0.15, 0.18, 0.20, 0.25]
sensitivities = [1]
especifities_1 = [1]

for t in threshold:
    df_probabilidades["prediccion"] = np.where(df_probabilidades[1]>=t, 1, 0)
    df_probabilidades["original"] = bank_data["y"]
    df_probabilidades
    
    confusion_matrix = pd.crosstab(df_probabilidades["prediccion"],df_probabilidades["original"])
    
    TN = confusion_matrix[0][0]
    TP = confusion_matrix[1][1]
    FN = confusion_matrix[1][0]
    FP = confusion_matrix[0][1]
    
    sens = TP/(TP+FN)
    sensitivities.append(sens)
    espc_1 = 1-TN/(TN+FP)
    especifities_1.append(espc_1)
    
sensitivities.append(0)
especifities_1.append(0)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.plot(especifities_1, sensitivities, marker="o", linestyle="--", color="r")
x=[i*0.01 for i in range (100)]
y=[i*0.01 for i in range (100)]
plt.plot(x,y)

plt.xlabel("especifity-1")
plt.ylabel("sensitivity")
plt.title("CURVA ROC")
```
