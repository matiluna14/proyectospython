```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### distribución uniforme (es decir todos los valores son igual de probables)

```python
a = 1
b = 100
n = 20000000

data = np.random.uniform(a,b,n)
```

```python
%matplotlib inline
plt.hist(data)
```

### distribución normal (es la más usada. Los mayoria de los valores están comprendidos entre -1 y 1. Los parámetros con los que cuenta esta distribución son la "media" y la "desviación típica". La particularidad de la distribución normal estandar es que la media es cero, es decir que la mayoría de los valores caen cerca del cero; y la desviación típica es -1 y 1)

```python
data = np.random.randn(100)
```

```python
x = range(0,100)
y = data
plt.plot(x,y)
```

```python
plt.hist(data)
```

### si quiero colocar mi propia media y mi propia desviación típica se hace de la sig. manera:

```python
media = 5
desviacionTipica = 2.5 ## 2.5 de desviación abajo y 2.5 de desviacón arriba
distribuciónNormalDeValores = np.random.randn(10000) ## 10000 números generados aleatoriamente que siguen una distribución normal.
data = media + desviacionTipica * distribuciónNormalDeValores  ## procede del teorema central del límite, y sirve para convertir cualquier valor aleatorio en una distribución normal
```

```python
plt.hist(data)
```

### Dummy Data Sets

```python
n=10000
data = pd.DataFrame(      ##la función "DataFrame" permite convertir listas de valores en dataframes
            {
                'A': np.random.randn(n),               ##genera una distribución normal tipica aleatoria de n valores.
                'B': 1.5 + 2.5 * np.random.randn(n),   ##genera una distribución normal aleatoria de n valores con media=1.5 y desviación=2.5
                'C': np.random.uniform(5,32,n)         ##genera una distribución uniforme aleatoria de n valores con límiteInferior=5 y limiteSuperior=32
            }
)
```
