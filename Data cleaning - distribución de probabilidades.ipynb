```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### distribuci√≥n uniforme (es decir todos los valores son igual de probables)

```python
a = 1
b = 100
n = 20000000

data = np.random.uniform(a,b,n)
```

```python
%matplotlib inline
plt.hist(data)
```

### distribuci√≥n normal (es la m√°s usada. Los mayoria de los valores est√°n comprendidos entre -1 y 1. Los par√°metros con los que cuenta esta distribuci√≥n son la "media" y la "desviaci√≥n t√≠pica". La particularidad de la distribuci√≥n normal estandar es que la media es cero, es decir que la mayor√≠a de los valores caen cerca del cero; y la desviaci√≥n t√≠pica es -1 y 1)

```python
data = np.random.randn(100)
```

```python
x = range(0,100)
y = data
plt.plot(x,y)
```

```python
plt.hist(data)
```

### si quiero colocar mi propia media y mi propia desviaci√≥n t√≠pica se hace de la sig. manera:

```python
media = 5
desviacionTipica = 2.5 ## 2.5 de desviaci√≥n abajo y 2.5 de desviac√≥n arriba
distribuci√≥nNormalDeValores = np.random.randn(10000) ## 10000 n√∫meros generados aleatoriamente que siguen una distribuci√≥n normal.
data = media + desviacionTipica * distribuci√≥nNormalDeValores  ## procede del teorema central del l√≠mite, y sirve para convertir cualquier valor aleatorio en una distribuci√≥n normal
```

```python
plt.hist(data)
```

### Dummy Data Sets

```python
n=10000
data = pd.DataFrame(      ##la funci√≥n "DataFrame" permite convertir listas de valores en dataframes
            {
                'A': np.random.randn(n),               ##genera una distribuci√≥n normal tipica aleatoria de n valores.
                'B': 1.5 + 2.5 * np.random.randn(n),   ##genera una distribuci√≥n normal aleatoria de n valores con media=1.5 y desviaci√≥n=2.5
                'C': np.random.uniform(5,32,n)         ##genera una distribuci√≥n uniforme aleatoria de n valores con l√≠miteInferior=5 y limiteSuperior=32
            }
)
```
